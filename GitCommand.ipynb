```python
# git command
# git 版本控制文件都在.git文件夹中

# git init && config  file
'''
git init
初始化代码仓库

watch -n 1 -d find .
每隔一秒刷新一次目录

ls -la

rm -rf .git/hooks/*.sample
删除代码仓库hooks文件夹下面.sample格式的文件

cat .git/config
查看本地config文件

cat ~/.gitconfig
当前用户目录下的gitconfig，存放global的文件

git config --global -l
打印global的配置

git config user.name "Kirin"
git config user.email "password"
优先使用本地信息commit

git config -l 
列出所有配置

git config --global -l
如果没有配置本地信息，使用global配置
'''

# git add && hash && blob object
''' 
# 工作区不归git管理
echo "hello git">hello.txt
cat hello.txt
>>>hello git

git status
>>> untracked files
        hello.txt

#tree
tree .git

# git add
git add hello.txt
提交到缓存区,objects中有文件生成，8d\index

git cat-file
查看objects里的文件(sha1 (secure hash agorithm))

git cat-file -t
查看文件的类型

git cat-file -t directory_name + sha1 hash value
# example:git cat-file -t 8d0e41
>>>blob

git cat-file -p
查看文件的内容

git cat-file -s
>>>10

git add 生成文件的对象，objects文件夹中只存储内容，不存储文件名
# 如果两个文件内容完全一样，只是文件名不同，git add之后objects文件夹不会生成新的blob文件

git知道哪些文件在工作区哪些在缓存区
'''

# hash agorithm
# hash generate index depend on input(content)
'''
把任意长度的输入通过散列算法变换成固定长度的输出
MD5 128bit\SHA1 160bit\SHA256 256bit\SHA512 512bit

SHA1 value format:长度40，16进制

echo "hello git" | shasum
>>> file content hash value 

echo "blob 10\0 hello git" | shasum
>>> git hash value

ls -lh
>>> size 10B
使用echo创建的文件有一个换行，所以10B，not 9B

tips:hash value与git add之后objects文件夹中blob文件的hash value不同
        说明git计算hash value不仅仅是文件内容的计算

git将原始文件进行压缩，存储到objects文件夹中
'''

# index area
# working directory->staging area(index)->git repository
'''
echo "file1" > file1.txt
git add file1.txt
tree .git/

cat ls-files
将当前在索引区里的文件显示出来
cat ls-files -s
列出文件的详细信息
>>>100644 e21297.......... 0 file1.txt
>>>文件权限+文件对应的blob对象+0+filename

index连接工作区和代码仓库

echo "file2" > file2.txt
# git ls_files -s 查看index里的文件

code file1.txt
git status
>>> modified:file1.txt
# 使用blob对象对比判断出工作区和缓存区不一致

git add file1.txt
# 重新生成一个新的对象
'''

# git commit
'''
git commit -m "message"

root commit 第一次commit

commit command generate commit object

git cat-file -t hashcode
commit

git cat-file -p shaonecode
tree hashcode

tree object代表目录结构的概念

git commit change .git/logs 

cat .git/HEAD
可以将HEAD文件视为一个指针，HEAD文件永远指向当前工作的分支

cat .git/refs/heads/master
当前工作的分支指向最近一次的commit object

git log 
查看记录

'''

#git branch
'''
git branch
cat .git/HEAD

# create a branch
git branch branch-name
git branch dev
dev和master指向同一commit文件

git checkout dev
转向dev分支

# delete a branch
git branch --delete dev

git branch -D dev
只删除了dev指针，但是commit对象还在，只不过无法访问了，所以是一个垃圾对象

'''
# checkout特定的commit
'''
detached head

git reflog

'''

#添加远程仓库
'''
tree .git
cat .git/refs/remotes/origin/master
cat .git/refs/heads/master
两个都指向了同一个commit对象
'''
```
